```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('/content/drive/MyDrive/Colab Notebooks/addsalt.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread("/content/drive/MyDrive/Colab Notebooks/gaussian_noise.jpg", cv2.IMREAD_GRAYSCALE)
img3 = cv2.imread("/content/drive/MyDrive/Colab Notebooks/speckle_noise.jpg", cv2.IMREAD_GRAYSCALE)

# Apply 3x3 Gaussian filter to reduce noise
salt = cv2.GaussianBlur(img1, (3, 3), 0)
gaussian = cv2.GaussianBlur(img2, (3, 3), 0)
speckle = cv2.GaussianBlur(img3, (3, 3), 0)

#Print
plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
plt.imshow(salt, cmap='gray')


plt.subplot(1, 3, 2)
plt.imshow(gaussian, cmap='gray')


plt.subplot(1, 3, 3)
plt.imshow(speckle, cmap='gray')

# Save the noisy and filtered images
cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/pics/deblursalt.jpg', img1)
cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/pics/deblursgaussian.jpg', img2)
cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/pics/deblurspeckle.jpg', img3)
```
